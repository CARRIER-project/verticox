```python
import numpy as np
import pandas as pd
from sksurv.datasets import load_whas500
from sksurv.linear_model import CoxPHSurvivalAnalysis
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
LIMIT=10
```

```python
X, y = load_whas500()

display(X[:5])
display(y[:5])
display(f'Number of features: {len(X.columns)}')
display(f'Number of samples" {X.shape[0]}')
```

```python
@np.vectorize
def include(event):
    return event[0]

non_censored_X = X[include(y)]
non_censored_y = y[include(y)]

non_censored_X
```

```python
cox_model = CoxPHSurvivalAnalysis()

cox_model.fit(non_censored_X, non_censored_y)
```
