```python
joined = df.join(parameter_df[["category"]])
joined
```

```python
joined["convergence_hours"] = joined["convergence_runtime"]/3600
```

## Scalability of parties and iterations


```python
parties_vs_iterations = joined[joined.category == "parties_vs_iterations"]
parties_vs_iterations
```

```python
parties_vs_iterations.plot("iterations", "convergence_runtime", "scatter")
```

```python
fig, ax = plt.subplots(3)

cmap = plt.get_cmap()

metrics = ["mse", "mad", "sad"]

def get_color(metric):
    x = metrics.index(metric)/(len(metrics)-1)
    return cmap(x)


for metricidx, metric in enumerate(["mse", "mad", "sad"]):
    for partyidx, party in enumerate(range(2, 6)):    
        subset = parties_vs_iterations[parties_vs_iterations.parties == party]
        ax[metricidx].scatter(subset["iterations"], subset[metric], color=get_color(metric), label=f"{party} parties")
        
plt.legend()


```

```python
parties_vs_iterations = parties_vs_iterations.sort_values(["parties", "iterations"])
```

```python
parties_vs_iterations[["parties", "iterations", "preparation_runtime", "convergence_runtime", "mse", "sad", "mad"]]\
    .set_index(["parties", "iterations"])\
    .style.format("{:.4e}", subset=["mse", "sad", "mad"])
```
