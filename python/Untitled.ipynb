```python
import json
```

```python
lines = None
with open("online.txt", "r") as f:
    lines = f.readlines()
```

```python
online = set()
for l in lines:
    if not l == "\n":
        try:
            l = l.replace("'", "\"")
            data = json.loads(l)["data"]
            for node in data:
                id_ = node['id']
                status = node['status']
                print(f"{id_}: {status}")
                if status == "online":
                    online.add(id_)
        except Exception:
            print(f"Exception, line:\n{l}")
        print("--------------")
        
                  
        
print(f"Have been online: {online}")
```
