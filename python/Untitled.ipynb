```python
for c in df.columns:
    if df[c].dtype == object:
        unique_values = df[c].unique()
        print(f"{c}: {unique_values}")
```

```python
for c in df.columns:
    if df[c].dtype == object:
        df[c] = df[c].astype("category")
        
df.head()
```

```python
df = df.drop(columns=["Unnamed: 3"])
```

```python
outcome = df[["Survival Months", "Status"]]
```

```python
df = df.drop(columns=outcome.columns)
```
