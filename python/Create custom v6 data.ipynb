```python
from sksurv import datasets
import numpy as np
import pandas as pd
```

```python
features, outcome = datasets.load_veterans_lung_cancer()
```

```python
features
```

```python
outcome
```

```python
# Introduce missing values
RATIO_MISSING = 0.1
num_records = len(features)
num_missing = round(RATIO_MISSING * num_records)
selection = np.random.choice(np.arange(num_records), size=num_missing, replace=False)
selection
```

```python
features.loc[selection, "Age_in_years"] = np.nan
features.loc[selection, "Celltype"] = np.nan
```

```python
features
```

```python
node1_df = features[features.columns[:3]]
node2_df = features[features.columns[3:]]
```

```python
event_happened = [o[0] for o in outcome]
event_time = [o[1] for o in outcome]

outcome_df = pd.DataFrame({"event_time": event_time, "event_happened": event_happened})

outcome_df
```
