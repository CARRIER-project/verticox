```python
import vantage6.client as v6client
import pandas as pd
import numpy as np
```

```python
from sksurv.datasets import load_whas500
```

```python
features, events = load_whas500()
feature_columns = features.columns
feature_columns
```

```python
event_happened = [e[0] for e in events]
event_time = [e[1] for e in events]

features['event_time'] = pd.Series(event_time)
features['event_happened'] = pd.Series(event_happened)

event_columns = ['event_time', 'event_happened']
```

```python
split = len(feature_columns)//2
split
```

```python
node1_features = feature_columns[:split]
node2_features = feature_columns[split:]
```

```python
node1_data = features[node1_features.to_list() + event_columns]
node2_data = features[node2_features.to_list() +  event_columns]
node3_data = features[event_columns]
display(node1_data)
display(node2_data)
display(node3_data)
```

```python
node1_data.to_csv("node1_whas500.csv", index=False)
node2_data.to_csv("node2_whas500.csv", index=False)
node3_data.to_csv("node3_whas500.csv", index=False)
```
